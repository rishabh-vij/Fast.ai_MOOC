## Training using transfer learning
**We'll use resnet34, which is trained on imagenet as our pretained model**

```python
from planet import f2

metrics=[f2]
f_model = resnet34
```

```python
labels=f'{PATH}train_v2.csv'
n=len(list(open(labels)))-1
val_idxs=get_cv_idxs(n)
```

**We define a function to process data at required sizes, because training our model at different sizes helps reduce the chances of overfitting**

```python
def get_data(sz):
    tfms=tfms_from_model(f_model=f_model,sz=sz,aug_tfms=transforms_top_down,max_zoom=1.05)
    dat= ImageClassifierData.from_csv(PATH,'train-jpg',val_idxs=val_idxs,tfms=tfms,
                                       csv_fname=labels,suffix='.jpg')
    return dat
```

```python
data=get_data(64)
```

```python
learn=ConvLearner.pretrained(f_model,data,metrics=metrics)
```

```python
lrf=learn.lr_find()
```

```python
learn.sched.plot()
```

```python
lr=0.2
```

```python
learn.fit(lr,3,cycle_len=1,cycle_mult=2)
```

## Differential learning rates
**We define lrs as we'll use different learning rates to train different groups of layers in our model. This will help improve accuracy**


**Note: As the satelite images are not close to the imagenet data, we keep learning rates for first two group of layers high, i.e. lr/9 and lr/3. If our satelite data would have been like the imagenet data we would have wanted to keep our learning rates too low. In that case the learning rate for first two groupsof layers would have been lr/1000 and lr/100**

```python
lrs=[lr/9,lr/3,lr]
```

```python
learn.unfreeze()
learn.fit(lrs,3,cycle_len=1,cycle_mult=2)
```

```python
data=get_data(128)
learn.freeze()
learn.fit(lr,3,cycle_len=1,cycle_mult=2)
```

```python
learn.unfreeze()
learn.fit(lrs,3,cycle_len=1,cycle_mult=2)
```

```python
data=get_data(256)
learn.freeze()
learn.fit(lr,3,cycle_len=1,cycle_mult=2)
```

```python
learn.unfreeze()
learn.fit(lrs,3,cycle_len=1,cycle_mult=2)
```

```python
multi_preds,y=learn.TTA()
preds=np.mean(multi_preds,0)
```
